#WYKRESY

```python
bins = [0, 1e+03, 1e+04,1e+05,1e+07,np.inf]
names = ['sladowe', 'słabe', 'srednie', 'silne', 'bardzo silne']
data['EnergyRange'] = pd.cut(data['ENG'], bins, labels=names)
```

```python
plt.figure(figsize=(8,8),dpi=100)
data['EnergyRange'].hist()
plt.title('Ilosc wstrząsów o danej energii')
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
data_minus_0 = data[data.Z < 0]
```

```python
fig = plt.figure(figsize=(10,8),dpi=100)
ax = plt.axes(projection='3d')

energy_data=data_minus_0[data_minus_0['EnergyRange']=='sladowe']
zdata = energy_data['Z']
xdata = energy_data ['X']
ydata = energy_data ['Y']

ax.scatter3D(xdata, ydata, zdata, c='blue',alpha=0.2,label='Śladowe' )

energy_data=data_minus_0[data_minus_0['EnergyRange']=='słabe']
zdata = energy_data['Z']
xdata = energy_data ['X']
ydata = energy_data ['Y']

ax.scatter3D(xdata, ydata, zdata, c='green',alpha=0.1,label='Słabe' )


energy_data=data_minus_0[data_minus_0['EnergyRange']=='srednie']
zdata = energy_data['Z']
xdata = energy_data ['X']
ydata = energy_data ['Y']

ax.scatter3D(xdata, ydata, zdata, c='orange',alpha=0.2,label='Średnie' )

energy_data=data_minus_0[data_minus_0['EnergyRange']=='silne']
zdata = energy_data['Z']
xdata = energy_data ['X']
ydata = energy_data ['Y']

ax.scatter3D(xdata, ydata, zdata, c='red',alpha=0.5,label='Silne' )

energy_data=data_minus_0[data_minus_0['EnergyRange']=='bardzo silne']
zdata = energy_data['Z']
xdata = energy_data ['X']
ydata = energy_data ['Y']

ax.scatter3D(xdata, ydata, zdata, c='black',alpha=1 ,label='Bardzo silne')
ax.legend()
plt.title('Rozłożenie przestrzenne wstrzasów')
plt.xlabel("X")
plt.ylabel("Y")
plt.zlabel("Z")

```

```python
fig = plt.figure(figsize=(10,8),dpi=100)
ax = plt.axes(projection='3d')


energy_data=data_minus_0[data_minus_0['EnergyRange']=='srednie']
zdata = energy_data['Z']
xdata = energy_data ['X']
ydata = energy_data ['Y']

ax.scatter3D(xdata, ydata, zdata, c='orange',alpha=0.2,label='Średnie' )

energy_data=data_minus_0[data_minus_0['EnergyRange']=='silne']
zdata = energy_data['Z']
xdata = energy_data ['X']
ydata = energy_data ['Y']

ax.scatter3D(xdata, ydata, zdata, c='red',alpha=0.5,label='Silne' )

energy_data=data_minus_0[data_minus_0['EnergyRange']=='bardzo silne']
zdata = energy_data['Z']
xdata = energy_data ['X']
ydata = energy_data ['Y']

ax.scatter3D(xdata, ydata, zdata, c='black',alpha=1 ,label='Bardzo silne')
ax.legend()
plt.title('Rozłożenie przestrzenne średnich-bardzo silnych wstrzasów')

```

```python
data_cat_counts=[ data_cat[x].value_counts().to_dict() for x in data_cat   ]
```

```python
plt.figure(figsize=(8,8),dpi=100)
plt.bar(*zip(*data_cat_counts[0].items())) 
plt.title('Ilosc wstrząsów w rejonach')
```

 Uwzgledniamy tylko oddziały w ktorych wystepuje wiecej niz 10 wstrzasow:

```python
for key in list(data_cat_counts[1].keys() ):
  if data_cat_counts[1][key] <10:
    del data_cat_counts[1][key]

```

```python
plt.figure(figsize=(8,8),dpi=100)
plt.bar(*zip(*data_cat_counts[1].items())) 
plt.title('Ilosc wstrząsów w poszczególnych oddziałach')
```
